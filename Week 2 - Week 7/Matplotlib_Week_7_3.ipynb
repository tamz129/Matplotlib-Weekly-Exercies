```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Sample dataset
n = np.linspace(-3.0,3.0,100)
x,y = np.meshgrid(n,n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(y.shape)
```

```python
#3D WireFrame
fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_wireframe(x,y,z, color = 'orange')
ax.set(xlabel = 'x',
       ylabel = 'y',
       zlabel = 'z',
       title = '3D Wireframe')

plt.show()
```

```python
#3D Surface Plot
fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_surface(x,y,z, color = 'orange')

ax.set(xlabel = 'x',
       ylabel = 'y',
       zlabel = 'z',
       title = '3D Surface Plot')

plt.show()
```
